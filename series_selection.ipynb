```python
import pandas as pd
import os
```

### accepted

```python
df = pd.read_csv("corflow/series_selection/accepted.csv", sep=";")
df
```

```python
len(df.ID.unique()), len(df.IMAGE_ID.unique()), len(df.SEGMENTATION_ID.unique()), len(df.studyFeedback_id.unique())
```

```python
filenames = os.listdir("corflow/series_selection")
dcm_files = [f for f in filenames if f.endswith("dcm")]
json_files = [f for f in filenames if f.endswith("json")]
len(json_files), len(dcm_files)
```

```python
[f for f in filenames if f not in dcm_files + json_files]
```

```python
json_image_ids = [f.replace(".json", "") for f in json_files]
dcm_image_ids = [f.replace(".dcm", "") for f in dcm_files]
```

```python
(
    len(set(df.IMAGE_ID) - set(json_image_ids)),
    len(set(json_image_ids) - set(df.IMAGE_ID))
)
```

```python
(
    len(set(df.IMAGE_ID) - set(dcm_image_ids)),
    len(set(dcm_image_ids) - set(df.IMAGE_ID))
)
```
