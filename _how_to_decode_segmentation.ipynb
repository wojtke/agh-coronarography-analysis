```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/frame_selection/dataset.csv")
df.head()
```

```python
some_seg = df["segmentation"][0]
some_seg
```

```python
import zlib
import numpy as np
import base64

decoded_data = base64.b64decode(some_seg)

decompressed_data = zlib.decompress(decoded_data)

height, width = 512, 512  
mask = np.frombuffer(decompressed_data, dtype=np.uint8).reshape(height, width)
mask
```

```python
plt.imshow(mask)
```
